```python
# Phase 1: Data Cleaning — Healthcare Dataset
# Dataset: Kaggle Healthcare Dataset by Prasad Patil
# Author: MAUREEN NDINDA WILLIAM
# Date: June 2026

import pandas as pd
import numpy as np
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Path to dataset in Google Drive
file_path = "/content/drive/MyDrive/HEALTHCARE /DATA/healthcare_dataset.csv"

# Load dataset
df = pd.read_csv(file_path)

# Display basic information
print(f"Shape: {df.shape}")
print(df.head())
```

```python
# ---- STEP 1: Drop Duplicates ----

print(f"Shape BEFORE dropping duplicates: {df.shape}")

df = df.drop_duplicates()
df = df.reset_index(drop=True)

print(f"Shape AFTER dropping duplicates: {df.shape}")
print(f"Rows removed: {55500 - df.shape[0]}")
```

```python
# ---- STEP 2: Fix Name Casing ----

print("BEFORE:")
print(df['Name'].head(5).tolist())

df['Name'] = df['Name'].str.strip().str.title()

print("\nAFTER:")
print(df['Name'].head(5).tolist())
```

```python
# ---- STEP 3: Convert Date Columns to Datetime ----

print("BEFORE:")
print(df[['Date of Admission', 'Discharge Date']].dtypes)

df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'])

print("\nAFTER:")
print(df[['Date of Admission', 'Discharge Date']].dtypes)

print("\nSample:")
print(df[['Date of Admission', 'Discharge Date']].head(3))
```

```python
# ---- STEP 4: Engineer Length of Stay ----

df['Length of Stay'] = (df['Discharge Date'] - df['Date of Admission']).dt.days

print("Sample:")
print(df[['Date of Admission', 'Discharge Date', 'Length of Stay']].head(5))

print(f"\nLength of Stay stats:")
print(df['Length of Stay'].describe())

print(f"\nAny zero or negative LOS: {(df['Length of Stay'] <= 0).sum()}")
```

```python
# ---- STEP 5: Remove Negative Billing Amounts ----

print(f"Rows BEFORE: {df.shape[0]}")
print(f"Negative/zero billing rows: {(df['Billing Amount'] <= 0).sum()}")

df = df[df['Billing Amount'] > 0]
df = df.reset_index(drop=True)

print(f"\nRows AFTER: {df.shape[0]}")
print(f"Rows removed: {54966 - df.shape[0]}")

print(f"\nBilling Amount stats after cleaning:")
print(df['Billing Amount'].describe().round(2))
```

```python
import os

# ---- STEP 6: Export Cleaned Dataset ----

# Define the output directory based on the original file path structure
output_dir = '/content/drive/MyDrive/HEALTHCARE /DATA/'
output_file_path = os.path.join(output_dir, 'healthcare_dataset_cleaned.csv')

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

df.to_csv(output_file_path, index=False)

print("✅ Cleaned dataset saved!")
print(f"Final shape: {df.shape}")
print(f"\nFinal column list:")
print(df.columns.tolist())
print(f"\nFinal dtypes:")
print(df.dtypes)
```
